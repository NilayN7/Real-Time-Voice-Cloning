```
from nltk.tokenize import sent_tokenize
with open('/content/gdrive/MyDrive/voice_cloning/andrew_huberman.txt', 'r') as f:
  data = f.read()
```

```
paragraphs = data.split("\n\n")
```

```
!pip install nltk
import nltk
nltk.download('punkt')
import re
paragraph_sentence_list = []
for paragraph in paragraphs:
    paragraph = paragraph.replace("\n", " ")
    paragraph = paragraph.replace(" — ", " ")
    paragraph = paragraph.replace(",", "")
    paragraph = re.sub(r'[^a-zA-Z0–9_*.,?!åäöèÅÄÖÈÉçëË]', ' ', paragraph)
    paragraph_sentence_list.append(sent_tokenize(paragraph))
```

```
print(len(paragraph_sentence_list[0]))
```

```
text = ""
count = 0
for paragraph in paragraph_sentence_list:
    if " ".join(paragraph[0]).isupper():
      with open("/content/gdrive/MyDrive/voice_cloning/text_files/transcript_"+str(count)+".txt", "w") as fw:
        fw.write(text)
        text = ""
        count += 1
        text += "\n".join(paragraph[0])
        text += "\n\n"
    else:
      text += "\n".join(paragraph[0])
      text += "\n\n"
```

```
import librosa
from pathlib import Path
file_name = Path("/content/Real-Time-Voice-CLoning/dataset_root/subfolder/folder_1/folder_2/folder_2/folder_3/Master_Sleep")
audio, sr = librosa.load(file_name)

# Get number of samples for 2 seconds; replace 2 by any number
buffer = 5 * sr

samples_total = len(audio)
samples_wrote = 0
counter = 1

while samples_wrote < samples_total:

    #check if the buffer is not exceeding total samples 
    if buffer > (samples_total - samples_wrote):
        buffer = samples_total - samples_wrote

    block = audio[samples_wrote : (samples_wrote + buffer)]
    out_filename = "split_" + str(counter) + "_" + file_name

    # Write 2 second segment
    librosa.output.write_wav(out_filename, block, sr)
    counter += 1
    samples_wrote += buffer
```
